```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from scipy.signal import convolve2d

gaussian_ker = np.matrix([
    [1,2,1],
    [2,4,2],
    [1,2,1]
])/16
derx_kernel = np.matrix([[1,-1]])
image = np.asarray(Image.open("./im2.jpeg"))
image = np.round(0.299*image[:, :, 0]+0.587 * image[:, :, 1]+0.114*image[:, :, 2]).astype(np.uint8)
plt.imshow(image,cmap='Grays')
plt.show()
smooth = convolve2d(image,gaussian_ker,mode='same')
plt.imshow(smooth,cmap='Grays')
plt.show()
derx = convolve2d(image,derx_kernel,mode='same')
plt.imshow(derx,cmap='Grays')
plt.show()
dery = convolve2d(image,derx_kernel.T,mode='same')
plt.imshow(dery,cmap='Grays')
plt.show()
plt.imshow(np.arctan(dery/derx),cmap='Grays')
plt.show()
dimage = np.zeros((image.shape[0]*3,image.shape[1]*3))
dimage[1::3,1::3] = np.arctan(dery/derx)
plt.imshow(dimage,cmap='Grays')
plt.show()
```
